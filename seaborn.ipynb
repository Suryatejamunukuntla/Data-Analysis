# Seaborn - Statistical Data Visualization

## What is seaborn?

Seaborn is python library that builds on the top of matplotlib that makes thee statistical data more easy and beautiful.

## Why we need it?

1. For creating plots like violin plot ,box plot and pair plots.

2. Automatically handles data from pandas Dataframes.

## Where we use it?

1. Data Analysis and Reporting.

2. Exploring relationship between Variables.

3. Machine Learning For visualizing datasets and results.

# 2.Import Seaborn

```python
import seaborn as sns
# Load Titanic dataset
df = sns.load_dataset("titanic")
df.head()
```

## 4. Histograms & Countplots

```python

# Histogram
import matplotlib.pylab as plt
sns.histplot(df['age'].dropna(), bins=10, kde=True)
plt.title("Histogram of Age")
plt.show()
 
# Countplot
sns.countplot(x='class', data=df)
plt.title("Count of Passengers by Class")
plt.show()
```

## 5. Scatter Plots

```python
## 5. Scatter Plots
sns.scatterplot(x='age', y='fare', data=df, hue='sex')
plt.title("Age vs Fare")
plt.show()
```

## 6. Boxplot & Violin Plot

```python
## 6. Boxplot & Violin Plot
# Boxplot
sns.boxplot(x='class', y='age', data=df)
plt.title("Boxplot of Age by Class")
plt.show()
 
# Violin plot
sns.violinplot(x='class', y='age', hue='sex', data=df, split=True)
plt.title("Violin Plot of Age by Class and Sex")
plt.show()
```

## 7. Pairplot

```python
### 7. Pairplot
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species')
plt.show()
```

## 8. Heatmap

```python
# Correlation heatmap (numeric columns only)
corr = iris.select_dtypes(include=['float64', 'int64']).corr()
 
import seaborn as sns
import matplotlib.pyplot as plt
 
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
```

## 9. Customization & Themes

```python
### 9. Customization & Themes
sns.set_theme(style="whitegrid")  # options: darkgrid, white, ticks, etc.
sns.histplot(df['age'].dropna(), bins=10, color='orange')
plt.title("Customized Histogram with Seaborn")
plt.show()
```

## 10. Bar Plot

```python
sns.barplot(x='class', y='fare', data=df, hue='sex')
plt.title("Average Fare by Class and Sex")
plt.show()
```

## Hands-On Exercise

### 1. Compare Passenger Survival by Class and Gender

```python

df = sns.load_dataset("titanic")
sns.barplot(x="class",y="survived",data=df,hue='sex')
plt.title("Passenger Survival by class and Gender")
plt.show()
```

### 2. Distribution of Passenger Ages

```python
sns.histplot(data=df, x="age", bins=30, kde=True)
plt.title("Distribution of Passenger Ages")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

### 3. Fare Distribution by Class

```python
sns.boxplot(data=df, x="class", y="fare")
plt.title("Fare Distribution by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()
```

### 4. Relationship Between Age and Fare

```python
sns.scatterplot(data=df, x="age", y="fare", hue="class",style='sex')
plt.title("Relationship Between Age and Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

### 5. Correlation Heatmap

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Titanic Dataset")
plt.show()
```
